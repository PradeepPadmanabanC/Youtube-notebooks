### Distance from a Point to a Hyperplane

Given:
- Hyperplane: $ w \cdot x + b = 0 $
- Point: $ x_0 \in \mathbb{R}^n $

**Steps:**

1. **Find the projection:**
   $$
   x = x_0 + \lambda w
   $$
   Substitute into the hyperplane equation:
   $$
   w \cdot (x_0 + \lambda w) + b = 0
   $$
   $$
   w \cdot x_0 + \lambda \|w\|^2 + b = 0
   $$
   $$
   \lambda = \frac{-(w \cdot x_0 + b)}{\|w\|^2}
   $$

3. **Closest point on the hyperplane:**
   $$
   x^* = x_0 + \lambda w 
        = x_0 - w \, \frac{w \cdot x_0 + b}{\|w\|^2}
   $$

4. **Distance calculation:**
   $$
   \text{Distance} = \|x_0 - x^*\|
                   = \left| \frac{w \cdot x_0 + b}{\|w\|} \right|
   $$

### Final Formula
$$
d = \frac{|w \cdot x_0 + b|}{\|w\|}
$$
